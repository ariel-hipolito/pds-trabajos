```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sig
import scipy.io as sio
import pandas as pd
from pydub import AudioSegment
```

### CARGA DE SEÑALES:

Se cargan señales fisiológicas (`ECG` y `PPG`) y señales de audio en formato `.wav` para su análisis espectral.


```python
# ECG y PPG
ecg_one_lead = np.load('ecg_sin_ruido.npy')
ppg = np.load('ppg_sin_ruido.npy')

fs_ecg = 1000  # Hz
fs_ppg = 400   # Hz

# Audios
fs_audio_A, wav_data_A = sio.wavfile.read('la cucaracha.wav')
fs_audio_B, wav_data_B = sio.wavfile.read('prueba psd.wav')
fs_audio_C, wav_data_C = sio.wavfile.read('silbido.wav')
```

### ANCHO DE BANDA: 

El ancho de banda de una señal es el **rango de frecuencias** donde se concentra la **mayor parte de su energía o potencia.** 

Para poder explicarla se pueden utilizar diferentes tipos de definiciones: 

- **CLASICA (-3dB)**: el ancho de banda es el intervalo de frecuencias donde la potencia no cae más de 3dB respecto del valor maximo. Es la más usada en sistemas lineales o filtros.
  
\begin{equation}
\text{BW} = f_2 - f_1 \quad \text{donde} \quad PSD(f) \geq PSD_{\text{max}} - 3\,\text{dB}
\end{equation}

- **POTENCIA ACUMULADA**: en señales reales como las biologicas, las de audio y demás, la energía no se corta bruscamente, sino que se distribuye. Por lo que se define el ancho de banda como el rango de frecuencias que **contiene un cierto porcentaje de la potencia total**. Buscando generalmenre la frecuencia hasta la cual se acumula el 90% de la energía del espectro (BW del 90%).

### CALCULO DEL ANCHO DE BANDA EFECTIVO:

Se define una función que determina la frecuencia hasta la cual se acumula el **90% de la potencia total**.

1) Calculo el paso de la frecuencia: $\Delta f = f[1] - f[0]$
   
3) Obtengo la potencia total de la señal: $\text{Potencia total de la señal:} \quad P_{\text{total}} = \sum_i PSD(f_i) \cdot \Delta f$

   
5) Obtengo la potencia acumulada: $\text{Potencia acumulada:} \quad P_{\text{acum}}(f) = \sum_{i=0}^{k} PSD(f_i) \cdot \Delta f$

   
7) Buscamos la frecuencia donde se alcanza el 90% de la potencia total: $\text{Frecuencia donde se alcanza el 90\% de la potencia:} \quad P_{\text{acum}}(f_{\text{BW}}) = 0.9 \cdot P_{\text{total}}$


```python
def obtener_bw(psd, ff, porcentaje=0.9):
    df = ff[1] - ff[0]
    pot_total = np.sum(psd) * df
    pot_acum = np.cumsum(psd) * df
    idx = np.where(pot_acum >= pot_total * porcentaje)[0][0]
    return pot_total, ff[idx]
```

### ESTIMACION DE LA DENSIDAD ESPECTRAL DE POTENCIA (PSD)

Se aplica el método de Welch a cada señal utilizando una ventana de Hann. Se podria aplicar cualquier otra ventana.  
El tamaño de cada segmento (`nperseg`) se elige proporcional a la longitud de la señal, haciendo un trabajo "artesano" para no perder resolucion espectral. 


```python
# ECG y PPG
ff_ecg, psd_ecg = sig.welch(ecg_one_lead, fs=fs_ecg, window='hann', nperseg=int(len(ecg_one_lead)/10))
ff_ppg, psd_ppg = sig.welch(ppg, fs=fs_ppg, window='hann', nperseg=int(len(ppg)/10))

# Audios
ff_A, psd_A = sig.welch(wav_data_A, fs=fs_audio_A, window='hann', nperseg=int(len(wav_data_A)/40))
ff_B, psd_B = sig.welch(wav_data_B, fs=fs_audio_B, window='hann', nperseg=int(len(wav_data_B)/40))
ff_C, psd_C = sig.welch(wav_data_C, fs=fs_audio_C, window='hann', nperseg=int(len(wav_data_C)/40))
```

### ANCHO DE BANDA EFECTIVO (90% DE LA POTENCIA)
Procedemos a calcular el BW en cada una de las señales cargadas

```python
pot_total_ecg, bw_ecg = obtener_bw(psd_ecg, ff_ecg)
pot_total_ppg, bw_ppg = obtener_bw(psd_ppg, ff_ppg)
pot_total_A, bw_A = obtener_bw(psd_A, ff_A)
pot_total_B, bw_B = obtener_bw(psd_B, ff_B)
pot_total_C, bw_C = obtener_bw(psd_C, ff_C)

tabla_bw = pd.DataFrame({
    'Señal': ['ECG', 'PPG', 'Audio A (La Cucaracha)', 'Audio B (Prueba)', 'Audio C (Silbido)'],
    'Frecuencia de muestreo [Hz]': [fs_ecg, fs_ppg, fs_audio_A, fs_audio_B, fs_audio_C],
    'Potencia total (integral PSD)': [pot_total_ecg, pot_total_ppg, pot_total_A, pot_total_B, pot_total_C],
    'BW (90% Potencia) [Hz]': [bw_ecg, bw_ppg, bw_A, bw_B, bw_C]
})
tabla_bw
```

## Gráficos de la PSD

Se representan las densidades espectrales de potencia de cada señal, con una línea verde que marca el ancho de banda que contiene el 90% de la potencia total.


```python
plt.figure(figsize=(18, 6))

def plot_psd(ff, psd, bw, title, xlim=None):
    psd_db = 10 * np.log10(psd / np.max(psd) + 1e-12)
    plt.plot(ff, psd_db)
    plt.axvline(bw, color='green', linestyle='--')
    plt.text(bw, -70, f'BW 90% = {bw:.1f} Hz', color='green', fontsize=9, rotation=0, va='bottom')
    plt.title(title)
    plt.xlabel("Frecuencia [Hz]"); plt.ylabel("PSD [dB]")
    plt.grid(True)
    if xlim:
        plt.xlim(xlim)
    plt.ylim([-80, 5])

# Subplots
plt.subplot(2, 3, 1); plot_psd(ff_ecg, psd_ecg, bw_ecg, "ECG")
plt.subplot(2, 3, 2); plot_psd(ff_ppg, psd_ppg, bw_ppg, "PPG")
plt.subplot(2, 3, 3); plot_psd(ff_A, psd_A, bw_A, "Audio A (La Cucaracha)", [0, 8000])
plt.subplot(2, 3, 4); plot_psd(ff_B, psd_B, bw_B, "Audio B (Prueba)", [0, 8000])
plt.subplot(2, 3, 5); plot_psd(ff_C, psd_C, bw_C, "Audio C (Silbido)", [0, 8000])
plt.tight_layout()
plt.show()
```

```python
#%% BONUS: SEÑAL PROPUESTA (SENO MODULADO)

# Cargar archivo MP3
audio = AudioSegment.from_mp3("whale_sound.mp3")

# Exportar a WAV
audio.export("whale_sound.wav", format="wav")

fs_ballena, senal_ballena = sio.wavfile.read('whale_sound.wav')

# Como es stereo lo convertiremos a mono tomando un canal
if senal_ballena.ndim > 1:
    senal_ballena = senal_ballena[:, 0]

# Estimación de PSD usando método de Welch
nperseg = int(len(senal_ballena) / 20)  # podés ajustar el denominador si querés
ff_ballena, psd_ballena = sig.welch(senal_ballena, fs=fs_ballena, window='hann', nperseg=nperseg)

# Cálculo del ancho de banda
bw_ballena = obtener_bw(psd_ballena, ff_ballena)
print(f"\nAncho de banda del sonido de ballena (90% potencia): {bw_ballena:.2f} Hz")

# Gráfico del espectro
plt.figure(figsize=(10, 4))
plt.plot(ff_ballena, 10 * np.log10(psd_ballena + 1e-12))
plt.axvline(bw_ballena, color='green', linestyle='--', label=f'BW 90% = {bw_ballena:.1f} Hz')
plt.title("PSD - Sonido de ballena (MP3 convertido a WAV)")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("PSD [dB]")
plt.legend()
plt.grid(True)
plt.show()
```
